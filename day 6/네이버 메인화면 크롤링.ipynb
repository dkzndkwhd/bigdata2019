```python
import urllib.request # 인터넷 연결 라이브러리
import bs4 # 분석 툴 라이브러리
```

```python
# !pip install bs4 -->라이브러리 인스톨시 사용하는 함수.
```

```python
url = "http://www.naver.com"
```

```python
html = urllib.request.urlopen(url) # 특정 사이트에 연결 . 받아왔을 뿐, 한번 더 함수를사용해주어야 한다.
```

```python
bs_obj = bs4.BeautifulSoup(html , "html.parser") # html문서로 인식하도록 분석해준다.
```

```python
# f12누르면 편하게 크롤링 가능 (사이트에서) 화살표 누르면 마우스 포인트로 직접 포인트 가능/.
```

```python
# 가장 앞에 있는건 태그 , . 뒤는 클래스 , 그 후가 아이디
```

```python
bs_obj.find("ul") # ul이라는 태그를 불러온다. ul은 태그명
```

```python
bs_obj.findAll("ul") # ul 은 모조리 가져온다
```

```python
ul = bs_obj.find("ul", {"class":"an_l"}) # ul 태그를 가진 것 중에서 클래스가 an_1인 것들을 가져오는 함수. 조건적으로
```

```python
# 태그와 태그 사이에 있는 것들을 가져오는 것
#ul>
#  <lib>
#  <lib>
#  <lib>
#/ul>
```

```python
ul
```

```python
lis = ul.findAll("li") # 리스트형태로 li태그 안에 있는 것들을 가져온다.
```

```python
lis
```

```python
for li in lis:
    print(li)
```

```python
len(lis)
```

```python
for li in lis:  # lis 리스트 안을 원소 하나씩 돌면서 원소 하나당 a 태그의 내용들을 긁어오는 것
    a_tag = li.find("a")
    print(a_tag)
```

```python
for li in lis: # a 태그 안의 span 클래스들만 가져오는 것.
    a_tag = li.find("a")
    span = a_tag.find("span")
    print(span)
```

```python
# a 태그 안에 span 클래스가 2개가 있었다. 뒤에꺼만 긁어오자
```

```python
for li in lis:
    a_tag = li.find("a")
    span = a_tag.find("span" , {"class" : "an_txt"})
    print(span.text) # 클래스 사이에 있는 text 가져올 때는 .text 를  사용
```
