```python
pop_seoul = pd.read_excel('../../Data/pop_in_seoul.xls', 
                          header = 2,
                          usecols = 'B, D, G, J, N',
                          encoding='utf-8') # 엑셀의 셀 이름별로 자기가 원하는 데이터를 선별적으로 가져올 수 있다. 헤더는 몇번째 행부터 가져올지(가로)
```

```python
pop_seoul.rename(columns={pop_seoul.columns[0]:'구별',
                          pop_seoul.columns[1]:'인구수',
                          pop_seoul.columns[2]:'한국인',
                          pop_seoul.columns[3]:'외국인',
                          pop_seoul.columns[4]:'고령자',
                         } , inplace = True)
```

```python
pop_seoul['외국인 비율'] = pop_seoul['외국인']/pop_seoul['인구수'] *100 # 파생변수 만들어서 넣기
```

```python
pop_seoul['고령자 비율'] = pop_seoul['고령자']/pop_seoul['인구수'] *100
```

```python
CCTV_seoul['최근증가율'] = (CCTV_seoul['2016년']+CCTV_seoul['2015년']+CCTV_seoul['2015년'])/CCTV_seoul['2013년도 이전']*100
```

```python
CCTV_seoul
```

```python
# dataframe 합칠때 merge 함수를 사용한다. pandas 내에 있다/.
```

```python
data_rewsult =pd.merge(CCTV_seoul , pop_seoul, on = '구별') # 공통된 부분을 기준으로 붙인다
```

```python
data_rewsult
```

```python
del data_rewsult['2013년도 이전']#열을 지워보자
del data_rewsult['2014년']
del data_rewsult['2015년']
del data_rewsult['2016년']
# 데이터 프레임의 컬럼 삭제 del
data_rewsult
```

```python
data_rewsult.set_index('구별', inplace = True) #구별로 구분해서 인덱싱
```
